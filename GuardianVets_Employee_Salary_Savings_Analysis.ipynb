```
df2025 = pd.read_csv("2025 (1).csv")
```

```
import pandas as pd

# Concatenating the two DataFrames
df_combined = pd.concat([df2024, df2025], ignore_index=True)

# Display the result
df_combined
```

```
df_emergency = df_combined[df_combined["call_outcome"] == "emergency"]
df_emergency
```

```
df_administrative = df_combined[df_combined["call_outcome"] == "administrative"]
df_administrative
```

```
df_appointment = df_combined[df_combined["call_outcome"] == "appointment"]
df_appointment
```

```
import pandas as pd

# Ensure datetime format
df_combined["call_start_time"] = pd.to_datetime(df_combined["call_start_time"])
df_combined["call_end_time"] = pd.to_datetime(df_combined["call_end_time"])

# Calculate call duration in seconds
df_combined["call_duration"] = (df_combined["call_end_time"] - df_combined["call_start_time"]).dt.total_seconds()

# Display the updated DataFrame
print(df_combined[["call_start_time", "call_end_time", "call_duration"]])
```

```
df_combined["call_duration"].mean()
```

```
import pandas as pd

# Load the data (make sure the file is uploaded)
df2024 = pd.read_csv("2024 (1).csv")

# Filter for rows that contain "VSCR" in the 'services' column
df_vscr = df2024[df2024["services"].astype(str).str.contains("VSCR", case=False)]

# Convert call_start_time and call_end_time to datetime
df_vscr["call_start_time"] = pd.to_datetime(df_vscr["call_start_time"])
df_vscr["call_end_time"] = pd.to_datetime(df_vscr["call_end_time"])

# Calculate call duration in seconds
df_vscr["call_duration"] = (df_vscr["call_end_time"] - df_vscr["call_start_time"]).dt.total_seconds()

# Display filtered and updated DataFrame
print(df_vscr[["case_id", "services", "call_start_time", "call_end_time", "call_duration"]])
```

```
df_vscr["call_duration"].mean()
```

```
import pandas as pd

# Load the data (make sure the file is uploaded)
df2024 = pd.read_csv("2024 (1).csv")

# Filter for rows that contain "After hours on call" in the 'services' column
df_after_hours = df2024[df2024["services"].astype(str).str.contains("After hours on call", case=False)]

# Convert call_start_time and call_end_time to datetime
df_after_hours["call_start_time"] = pd.to_datetime(df_after_hours["call_start_time"])
df_after_hours["call_end_time"] = pd.to_datetime(df_after_hours["call_end_time"])

# Calculate call duration in seconds
df_after_hours["call_duration"] = (df_after_hours["call_end_time"] - df_after_hours["call_start_time"]).dt.total_seconds()

# Display filtered and updated DataFrame
print(df_after_hours[["case_id", "services", "call_start_time", "call_end_time", "call_duration"]])
```

```
df_after_hours["call_duration"].mean()
```

```
import pandas as pd

# Load the data
df2024 = pd.read_csv("2024 (1).csv")

# Ensure the 'services' column is treated as string
services_col = df2024["services"].astype(str)

# Count rows containing "VSCR"
vscr_count = services_col.str.contains("VSCR", case=False).sum()

# Count rows containing "After hours on call"
after_hours_count = services_col.str.contains("After hours on call", case=False).sum()

# Print the results
print(f"Rows with 'VSCR': {vscr_count}")
print(f"Rows with 'After hours on call': {after_hours_count}")
```
